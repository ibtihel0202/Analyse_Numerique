### Correction exercice 1
## **1)**
*Avec Matlab √©valuer  ùëù(ùë•)  et  ùëì(ùë•)  pour  ùë•  allant de  .998  √†  1.002  en choisissant des pas de  .0001 .*

```python
def horner(a,x):
    n = len(a)
    result = a[1]
    for i in range(2,n):
        result = result * x + a[i]
        return result
    
```

```python
def p(x):
    return (x-1)**6
```

```python
def f(x):
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

```python
p(0.998)
```

```python
f(0.998)
```

```python
T=[1, -6, 15, -20, 15, -6, 1]
print("T= " , T)
print('-------- p(i) ---------')
tt=[0.998,0.001,1.002]
for i in tt :
 print("ans=" ,p(i))
print('-------- f(i) ---------')
tt=[0.998,0.001,1.002]
for i in tt :
 print("ans=" ,f(i))
print('-------- horner ---------')
tt=[0.998,0.001,1.002]
for i in tt :
 print("ans=" ,horner(T,i))
```
