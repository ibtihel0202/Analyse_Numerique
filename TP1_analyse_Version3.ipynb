```python
def horner(a,x):
    n = len(a)
    result = a[1]
    for i in range(2,n):
        result = result * x + a[i]
        return result
    
```

```python
def p(x):
    return (x-1)**6
```

```python
def f(x):
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

# Correction exercice 4
**1)**

*On consid√®re la fonction  ùëì(ùë•)=ùëí‚àíùë•‚àíùë• . Ecrire puis appliquer une fonction Matlab que l'on appellera newton.m √† la >fonction  ùëì , en prenant  ùë•0=0 ,  ùëíùëùùë†=0.001  .*

```python
def newtonn(a, eps):
    x1= a
    x2= x1-f(x1)/fprime(x1)
    while abs(x1 - x2) > eps:
        x1 = x2
        x2 = x1 - f(x1) / fprime(x1)
    return  x2
```
