```python
def horner(a,x):
    n = len(a)
    result = a[1]
    for i in range(2,n):
        result = result * x + a[i]
        return result
    
```

```python
def p(x):
    return (x-1)**6
```

```python
def f(x):
    return x**6-6*x**5+15*x**4-20*x**3+15*x**2-6*x+1
```

*Utiliser un programme dicho.m qui permettra de trouver les solutions de l'√©quation  ùëì(ùë•)=0  avec  ùëíùëùùë†=0.001 .*

```python
def dicho (f ,a ,b , eps ):

    g , d = min (a , b ) , max (a , b )
    valg , vald = f ( g ) , f ( d )
    n = 0
    while d - g > eps :
        n += 1
        m = ( g + d )/2
        valm = f ( m )
        if valg * valm < 0 :
            d = m
            vald = valm
        else :
            g = m
            valg = valm
    print (" nombre d'it√©rations n√©cessaires :", n )
    return ( g + d )/2
```

**1)**

*On consid√®re la fonction  ùëì(ùë•)=ùëí‚àíùë•‚àíùë• . Ecrire puis appliquer une fonction Matlab que l'on appellera newton.m √† la >fonction  ùëì , en prenant  ùë•0=0 ,  ùëíùëùùë†=0.001  .*

```python
def newton(a, eps):
    x1= a
    x2= x1-f(x1)/f_prime(x1)
    while abs(x1 - x2) > eps:
        x1 = x2
        x2 = x1 - f(x1) / f_prime(x1)
    return  x2
```

```python
import math
f= lambda x:math.exp(-x)-x
f_prime =lambda x:-math.exp(-x)-1
```

```python
sol = newton(0, 0.001)
print(sol)
```
